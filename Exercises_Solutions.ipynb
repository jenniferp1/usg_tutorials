```python

interesting_refs=["reference_1.txt","reference_4.txt","reference_5.txt"]
```

```python
cwd = os.getcwd()
ref_path = cwd+"/demo_references"
```

```python

entities = pywand.gettext(ref_path, interesting_refs)
```

```python

for i in range(len(entities)):
    print(entities[i])
```

```python

for i in range(len(entities)):
    for entity,count in entities[i]:
        print(entity,count,interesting_refs[i])
```

```python

df_entities = pd.DataFrame(columns=["name","ref"])
df_entities
```

```python

names = []
refs = []
```

```python

for i in range(len(entities)):
    for entity,count in entities[i]:
        names.append(entity)
        refs.append(interesting_refs[i])
```

```python

df_entities["name"] = names
df_entities["ref"] = refs
df_entities.head()
```
